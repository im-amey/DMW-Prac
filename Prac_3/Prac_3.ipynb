```python
import pandas as pd
```

```python
df = pd.read_csv('../DMW_DataSets/Online Retail.csv',encoding='latin1')
df.head()
```

```python
df_cleaned = df.dropna(subset=['Description', 'CustomerID']).copy()
```

```python
price_bins = [-float('inf'), 0, 5, 20, 50, 100, 500, 1000, float('inf')]

price_labels = ['Invalid', 'Free', 'Cheap', 'Affordable', 'Moderate', 'Expensive', 'Very Expensive', 'Luxury']
```

```python
df_cleaned.loc[:, 'PriceCategory'] = pd.cut(df_cleaned['UnitPrice'], bins=price_bins, labels=price_labels, include_lowest=True)
df_cleaned.head()
```

```python
qcut_quantity_labels = ['Very Low', 'Low', 'Medium', 'High']
df_cleaned.loc[:, 'QCutQualityCategory'] = pd.qcut(df_cleaned['Quantity'], q=4, labels=qcut_quantity_labels)
print(df_cleaned)
```

```python
product_hierarchy = {
    'WHITE HANGING HEART T-LIGHT HOLDER':'Home Decor',
    'REGENCY CAKESTAND 3 TIER':'Kitchenware',
    'RED RETROSPOT BIRD HOUSE':'Garden',
    'JUMBO BAG RED RETROSPOT':'Bag',
    'ASSORTED COLOUR BIRD ORNAMENT':'Home Decor'
}
```

```python
df_cleaned.loc[:,'ProductCategory'] = df_cleaned['Description'].map(product_hierarchy).fillna('Other')
```

```python
print('-----------------------------------')
print("\nUpdated Dataset with Discretization and Concept Hierarchy:")
print(df_cleaned.head())
```

```python
df_cleaned = df.dropna(subset=['Description', 'CustomerID']).copy()

price_bins = [-float('inf'), 0, 5, 20, 50, 100, 500, 1000, float('inf')]
price_labels = ['Invalid', 'Free', 'Cheap', 'Affordable', 'Moderate', 'Expensive', 'Very Expensive', 'Luxury']

df_cleaned.loc[:, 'PriceCategory'] = pd.cut(df_cleaned['UnitPrice'], bins=price_bins, labels=price_labels, include_lowest=True)
df_cleaned.head()

qcut_quantity_labels = ['Very Low', 'Low', 'Medium', 'High']
df_cleaned.loc[:, 'QCutQualityCategory'] = pd.qcut(df_cleaned['Quantity'], q=4, labels=qcut_quantity_labels)
print(df_cleaned)

product_hierarchy = {
    'WHITE HANGING HEART T-LIGHT HOLDER':'Home Decor',
    'REGENCY CAKESTAND 3 TIER':'Kitchenware',
    'RED RETROSPOT BIRD HOUSE':'Garden',
    'JUMBO BAG RED RETROSPOT':'Bag',
    'ASSORTED COLOUR BIRD ORNAMENT':'Home Decor'
}

df_cleaned.loc[:,'ProductCategory'] = df_cleaned['Description'].map(product_hierarchy).fillna('Other')

print('-----------------------------------')
print("\nUpdated Dataset with Discretization and Concept Hierarchy:")
print(df_cleaned.head())
```
