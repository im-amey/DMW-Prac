```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn import datasets
```

```python
iris = datasets.load_iris()
print(iris.DESCR)
```

```python
feature_names = ['sep_len', 'sep_wid', 'pet_len', 'pet_wid']
df = pd.DataFrame(iris.data, columns=feature_names)
df.head()
```

```python
lr = LinearRegression()
lr.fit(df[['pet_len']].values, df.pet_wid.values)
```

```python
plt.xlabel("Petal Length")
plt.ylabel("Petal Width")
plt.scatter(df.pet_len.values, df.pet_wid.values, color='red', marker='+')
plt.plot(df.pet_len.values, lr.predict(df[['pet_len']].values), color='blue')
plt.plot()
```
